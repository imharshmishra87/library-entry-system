```python
# designing the library scratch for linear regression
class scratch:
  def __init__(self):
    self.m=None
    self.b=None
  def fit(self, X_train, y_train):
    data=[]
    data2=[]
    for i in range(X_train.shape[0]):
        data.append((X_train.iloc[i]-X_train.mean())*(y_train.iloc[i]-y_train.mean()))
        data2.append((X_train.iloc[i]-X_train.mean())**2)
    self.m=sum(data)/sum(data2)
    self.b=y_train.mean()-m*X_train.mean()

  def predict(self,X_test):
    y=self.m*X_test+self.b
    return y
```
