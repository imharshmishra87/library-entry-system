#Gradient Descent Intitution & Mathematical Calculation

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
```

```python
X,y=make_regression(n_samples=4,n_features=1,n_targets=1,noise=50,random_state=16)
```

```python
lr=LinearRegression()
lr.fit(X,y)
```

```python
m=lr.coef_
m
```

```python
lr.intercept_
```

```python
# By using ols we got the exact value of b and m
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color='green')
```

```python
# applyinng gradient descent algorithm
# for b=0
y_pred1=(m*X)+0
```

```python
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(12,4))
ax1.scatter(X,y)
ax1.plot(X,lr.predict(X),color='green')
ax2.scatter(X,y)
ax2.plot(X,lr.predict(X),color='green')
ax2.plot(X,y_pred1,color='red')
```

```python
#iteration 1
# Calculating the value of the slope
b=0
slope=-2*np.sum(y-m*X.ravel()-b)
slope
```

```python
# calculating the step size n is the learning rate
n=0.1
step_size=slope*n
step_size
```

```python
#calculating the value of b
b=b-step_size
b
```

```python
y_pred2=((m*X)+b).reshape(4)
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(12,4))
ax1.scatter(X,y)
ax1.plot(X,lr.predict(X),color='green')
ax2.scatter(X,y)
ax2.plot(X,lr.predict(X),color='green')
ax2.plot(X,y_pred1,color='red')
ax2.plot(X,y_pred2,color='blue')
```

```python
# iteration 2
slope=-2*np.sum(y- m*X.ravel()-b)
step_size=n*slope
b=b-step_size
```

```python
y_pred3=(m*X + b).reshape(4)
plt.plot(X,lr.predict(X),color='black')
plt.plot(X,y_pred1, color='green')
plt.plot(X,y_pred2,color='blue')
plt.plot(X,y_pred3,color='red')
```

# Grdient Descent Class From scratch

```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
```

```python
X,y=make_regression(n_samples=100,n_features=1,n_targets=1,noise=20,n_informative=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1)
```

```python
X_test.shape
```

```python
lr=LinearRegression()
lr.fit(X,y)
lr.coef_
lr.intercept_
```

```python
class gdregresso:
  def __init__(self,learn_rate,epochs):
    self.n=learn_rate
    self.epochs=epochs
    self.m=100
    self.b=-120


  def fit(self,X,y):
    for i in range(self.epochs):
      slope=-2*np.sum(y-self.m*X.ravel()-self.b)
      slope_m=-2*np.sum((y-self.m*X.ravel()-self.b)*X.ravel())
      step_size=self.n*slope
      step_size_m=self.n*slope_m
      self.b=self.b-step_size
      self.m=self.m-step_size_m

  def predict(self,X):
    y=self.m*X+self.b
    return y

```

```python
gdr=gdregresso(0.001,75)
gdr.fit(X_train,y_train)
```

```python
y_predict=gdr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test,y_predict))
```
